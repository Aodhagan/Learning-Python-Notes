# Program Execution
## Python Interpreter
Python is a programming language but it is also an interpreter. A Python Interpreter is a software package or program that executes other programs.

## Programmers view of Python
Python code to the programmer is simply a text file with python code written in it e.g.,   


```python
print("hello world!")
print(2 ** 100)
```

## Python's View
When you instruct Python to run the above script there are a few steps that Python carries out before the code actually starts being executed. The first step is to convert the code to **byte code** and then routed to a "virtual machine".



### Byte Code Compilation
Python compiles your source code into a format known as byte code. Byte code is a lower-level platform independent representation of your source code that is Python specific (it is not binary machine code for Intel CPUs). Python translates each of your source statements into a group of byte code instructions by decomposing them into individual steps. This is done to increase the speed of execution.

Byte code is stored in `.pyc` files or or compiled `.py` source code.

After Python version 3.2 byte code is stored in a `__pycache__` subdirectory. This speeds up startup as the file does not need to be compiled if the program is executed again. Events that will trigger recompilation of my code will be
1. Changes to the source code
2. if the file must be recompiled because it was created by a different Python version

Python can run the program using just `.pyc` files - it does not need the source code.

### The Python Virtual Machine (PVM)
once the byte code has been loaded it is shipped off to the Python Virtual Machine for execution. The PVM is just a code loop that iterates through your byte code instructions one by one to carry out their operations. It is the runtime engine of Python. This means that the system that compiles the code is the same system that executes the source code.

## Python Implementation Alternatives
There are a number of different Python implementations all with their own purposes and implement the same Python language but execute programs in different ways. They include
- CPython
- Jython - targeted for integration with Java programming language - converting to Java byte code
- IronPython - integrated with Windows .NET framework
- Stackless - Python for concurrency/multiprocessing
- PyPy

CPython is the standard implementation. PyPy is a drop-in replacement for CPython but runs programs much faster.

## Frozen Binaries
It is possible to turn your Python programs into true executables such that they can be run without requiring a Python installation combining the byte code with the Python virtual machine.
